## Question 1

Define a procedure called "check_input" that takes an input as a square list of lists (2D Matrix of N x N) and returns a boolen "true" if the input is valid or "false" otherwise, according to the following criteria.

1. Each coloumn must contain numbers from between 1 and N, with no duplicates.
2. Each row must contain numbers from between 1 and N, with no duplicates.

```python
def check_input(square):
    # checking for data being an integer value
    for i in range(len(square)):
        for j in range(len(square)):
          if isinstance(square[i][j], int):
            continue
          else:
            return False

    # checking in rows
    for i in range(len(square)):
        val = square[i][0]
        for j in range(len(square[0])):
            if val == square[i][j] and j != 0:
                return False

    # checking for columns
    for i in range(len(square)):
        val = square[0][i]
        for j in range(len(square[0])):
            if val == square[j][i] and j != 0:
                return False

    return True
```

```python
# run this block to test code after defining the procedure

test1 = [[1, 2, 3],
        [2, 3, 1],
        [3, 1, 2]]

test2 = [[1, 2, 3, 4],
        [2, 3, 1, 3],
        [3, 1, 2, 3],
        [4, 4, 4, 4]]

test3 = [['a', 'b', 'c'],
        ['b', 'c', 'a'],
        ['c', 'a', 'b']]

print(check_input(test1)) # must return true
print(check_input(test2)) # must return false
print(check_input(test3)) # must return false
```

**Part C** - Create a linear regression model using sklearn and fit it to the data you just imported


```python
from sklearn import linear_model

num = int(len(newdata)*0.8)
train = newdata[:num] # 80% traning
test = newdata[num:]  #20% testing

print("data", len(newdata))
print("train", len(train))
print("test", len(test))

model = linear_model.LinearRegression()
train_x_axis = np.array(train[["Mass (kg)"]])
train_y_axis = np.array(train[["Tail Length (cm)"]])

model.fit(train_x_axis, train_y_axis)

slope = model.coef_
intercept = model.intercept_
```

**Part D** - Print the values for slope and y-intercept

```python
print("slope", slope[0])
print("intercept", intercept)
```

**Part E** - Plot a scatter graph of the original data and a regression line on the same figure (example with dummy data shown below)
